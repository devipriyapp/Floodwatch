```python
import cv2
import numpy as np

video = "/content/drive/MyDrive/InShot_20240222_210623843.mp4"

cap = cv2.VideoCapture(video)
print(cap)

pixel_counts = []

while cap.isOpened():
    print("Opened")
    success, frame = cap.read()
    print(success)
    if success:
        print("success.....")
        results = model(frame) # Ivide predict option try cheyynm...

        # Visualize the results on the frame
        annotated_frame = results[0].plot()

        #__________________________________________________________Pixel Area Found_________________________________________________
        # Save the annotated frame to a temporary file
        cv2.imwrite("temp_annotated_frame.jpg", annotated_frame)

        # Read the annotated frame from the temporary file
        image = cv2.imread("temp_annotated_frame.jpg")
        hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

        # Dark Region____________________________________
        lower_red_dark = np.array([0,100,100])
        upper_red_dark = np.array([20,255,255])
        #_______________________________________________
        # Light Region___________________________________
        lower_blue_light = np.array([160,100,100])
        upper_blue_light = np.array([180,255,255])
        #_______________________________________________

        mask_dark = cv2.inRange(hsv_image, lower_red_dark, upper_red_dark)
        mask_light = cv2.inRange(hsv_image, lower_blue_light, upper_blue_light)

        pixel_count_dark = cv2.countNonZero(mask_dark)
        pixel_count_light = cv2.countNonZero(mask_light)

        total_pixel_count = pixel_count_dark + pixel_count_light

        pixel_counts.append(total_pixel_count)

        print("Total Area = {} pixels".format(total_pixel_count))

        # Calculate the threshold (e.g., x% of the initial pixel value)
        x_percent = 10 # Adjust as needed
        threshold = x_percent / 100 * pixel_counts[0]

        # Check for flooding
        if total_pixel_count > threshold:
            print("Flooding detected! Pixel count ({}) exceeds {}% of initial value.".format(total_pixel_count, x_percent))
            break

        #________________________________________________________________________________________________________________________________

        #if cv2.waitKey(1) & 0xFF == ord("q"):
            #break
    else:
        break

cap.release()
```
